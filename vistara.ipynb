```python
df['genres[0]']= df['genres[0]'].fillna('')
df['genres[1]']= df['genres[1]'].fillna('')

df['description']= df['description'].fillna('')
df['cast[0]']= df['cast[0]'].fillna('')
df['cast[1]']= df['cast[1]'].fillna('')
df['cast[2]']= df['cast[2]'].fillna('')
```

```python
import ast
df['genres'] = df[['genres[0]', 'genres[1]']].fillna('').apply(
    lambda x: ' '.join(x).strip(), axis=1
)
```

```python
df['cast'] = df[['cast[0]', 'cast[1]', 'cast[2]']].fillna('').apply(
    lambda x: ' '.join(x).strip(), axis=1
)
```

```python
df['tags'] = df['description']+ " " + df['genres'] 

df.head()

```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
     
```

```python
def preprocess_text(text):
  text = str(text).lower()
  text = re.sub(r'[^a-zA-Z\s]', '', text)
  words = text.split()
  words = [word for word in words if word not in stop_words]
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['tags'])

tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 5):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story 5' ,1)

```
